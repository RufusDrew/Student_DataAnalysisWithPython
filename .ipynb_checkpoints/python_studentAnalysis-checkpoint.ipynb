```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_rec.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum() #to know which value is null 
```

```python
df = df.drop("Unnamed: 0", axis = 1) #to drop perticular collumn 
print(df.head())
```

#change weekly study hours 

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May", "5-10")
df.head()
```

# Gender distribution 

```python
plt.figure(figsize = (5,5)) #chart size
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0]) #to create label to show exact counnt
plt.title("Gender Distribution")
plt.show()
```

```python
# from this above we have analysed:
#that the number pf a femal is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": "mean", "WritingScore": "mean"})
print(gb)
```

```python
plt.figure(figsize = (4,4)) #chart size
sns.heatmap(gb, annot = True) #heatmap chart with value
plt.title("Relationship Between Parent's Education and Student Score's")
plt.show()
```

```python
# from the above chart we have conclude that:
#The education of the parents have a good impact on the scores 
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": "mean", "WritingScore": "mean"})
print(gb1)
```

```python
plt.figure(figsize = (4,4)) #chart size
sns.heatmap(gb1, annot = True) #heatmap chart with value
plt.title("Relationship Between Parent's Marital Status and Student Score's")
plt.show()
```

```python
# from the above chart we have conclude that:
#The Marital status of the parents have a no impact on the students scores 
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("The Distribution of Ethnic Groups")
plt.show()
```
